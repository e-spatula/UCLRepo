# Analysis of Credit Card and Boston Housing Data

```r
load("credit_card.RData")
load("boston_housing.RData")
load("day5_test_data.RData")

library(texreg)
library(corrplot)
library(MASS)
library(leaps)
```

# Boston Housing Data

```r
plot(medv ~ ., data = boston_housing)
```

We can now have a look at the correlations between variables to get an idea about which predictors have the strongest correlation with median house price (medv) and therefore will produce the highest R^2 values.

```r
corrplot(cor(boston_housing))
```

From this correlation diagram it would appear that medv is closely correlated with lstat, ptratio, rm, and nox. Let's produce a linear model with these predictors. 

```r
optimum_model  <- lm(medv ~ lstat + ptratio + rm + nox, data = boston_housing)
summary(optimum_model)
```

The R^2 is fairly respectable and all the predictors appear to have a statistically significant effect except nox, which we'll remove and then compare the model to the base data to check if the increase in predictive power is statistically significant. 

```r
optimum_model  <- lm(medv ~ lstat + ptratio + rm, data = boston_housing)
vanilla_model  <- lm(medv ~ 1, data = boston_housing)

anova(vanilla_model, optimum_model)
```

```r
lm.min = lm(medv ~ 1, data = boston_housing)
lm.max = lm(medv ~ ., data = boston_housing)

scp = list(lower = lm.min, upper = lm.max)
forward_model  <- stepAIC(lm.min, 
                         scope = scp,
                         direction = "forward",
                         steps = 20)

backward_model  <- stepAIC(lm.max, 
                         scope = scp,
                         direction = "backward",
                         steps = 20)
both_model  <- stepAIC(lm.min, 
                         scope = scp,
                         direction = "both",
                         steps = 20)

```

```r
summary(forward_model)
summary(backward_model)
summary(both_model)
```

So now we have our step-wise selection models let's have a go at an exhaustive search.

```r
regsubsets.out  <- regsubsets(medv ~ .,
                             data = boston_housing,
                             nbest = 1, 
                              nvmax = NULL,
                             force.in = NULL,
                             force.out = NULL,
                             method = "exhaustive")


best.r2  <- max(summary(regsubsets.out)$adjr2)
idx  <- which(summary(regsubsets.out)$adjr2 == best.r2)
best.r2.predictors  <- summary(regsubsets.out)$outmat[idx,]
print(best.r2.predictors)
print(best.r2)

# convert the exhaustive model to a linear model that can be tested.

exhaustive_model  <- lm(medv ~ crim + zn + chas + nox + rm + dis + tax + black + lstat, data = boston_housing)
```

Now we have our exhaustive-search model let's try them all out against some test data. 

```r
boston_housing_test$forward_pred  <- predict(forward_model, newdata = boston_housing_test)
boston_housing_test$backward_pred  <- predict(backward_model, newdata = boston_housing_test)
boston_housing_test$both_pred  <- predict(both_model, newdata = boston_housing_test)
boston_housing_test$exhaustive_pred  <- predict(exhaustive_model, newdata = boston_housing_test)

adj_r_sq  <- function(predicted_column, actual_column) {
    
    actual_mean  <- mean(actual_column)
    squared_deviation  <- (actual_column - actual_mean)^2
    squared_residual  <- (actual_column - predicted_column)^2
    
    TSS  <- sum(squared_deviation)
    RSS  <- sum(squared_residual)
    Rsq  <- 1 - RSS/TSS
    
    return(Rsq)
}
attach(boston_housing_test)
adj_r_sq(forward_pred, medv)
adj_r_sq(backward_pred, medv)
adj_r_sq(both_pred, medv)
adj_r_sq(exhaustive_pred, medv)
detach(boston_housing_test)
```
